```python
from scipy import stats
def remove_outliers_zscore(df):
    outliers = {}
    for col in df.columns:
        if df[col].dtype != "object":
            z_scores = np.abs(stats.zscore(df[col]))
            threshold = 3
            outliers[col] = df[col][z_scores > threshold].tolist()
    return outliers

def clean_data_zscore(df):
    outliers = remove_outliers_zscore(df)
    for feature, values in outliers.items():
        df = df[~df[feature].isin(values)]
    return df
```

```python
# Create a BinaryEncoder object
encoder = BinaryEncoder()

# Initialize an empty DataFrame to store encoded features
encoded_df = pd.DataFrame()

# Iterate over each column in the DataFrame
for col in df.columns:
    # Check if the column is categorical (object dtype)
    if df[col].dtype == 'object' and  col!='Attrition':
        # Fit-transform the column using BinaryEncoder and store the result in a new DataFrame
        encoded_col = encoder.fit_transform(df[col])
        # Concatenate the encoded column with the previously encoded columns
        encoded_df = pd.concat([encoded_df, encoded_col], axis=1)
    else:
        # If the column is not categorical, simply concatenate it with the encoded DataFrame
        encoded_df = pd.concat([encoded_df, df[col]], axis=1)

# Display the encoded DataFrame
print(encoded_df)

        
```

```python
#Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=0)
```

```python
def find_high_correlated_features(data,threshold):
    features_to_remove=set()
    cor=data.corr()
    for i in range(len(cor.columns)):
        for j in range(i):
            if abs(cor.iloc[i,j])>threshold:
                features_to_remove.add(cor.columns[i])
    return features_to_remove
                
                
```

```python
drop_features=find_high_correlated_features(x_train,0.75)
drop_features
```

```python
x1_train=x_train.drop(list(drop_features),axis=1)
```

```python
x1_test=x_test[x1_train.columns]
y1_train=y_train
y1_test=y_test
```

```python
#Let perform the mutual information feature selection
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(x_train,y_train)
mutual_info
```

```python
from sklearn.feature_selection import SelectKBest
sel_ten_cols=SelectKBest(mutual_info_classif,k=15)
sel_ten_cols.fit(x_train,y_train)
a=x_train.columns[sel_ten_cols.get_support()]
a
```

```python
x2_train=x_train[a]
```

```python
x2_test=x_test[x2_train.columns]
y2_train=y_train
y2_test=y_test
```

```python
## Lets us do RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Initialize a logistic regression model (replace this with your chosen model)
model = LogisticRegression()

# Initialize RFE with the chosen model and desired number of features to keep
n_features_to_select = 15
rfe = RFE(model, n_features_to_select=15)

# Fit RFE to the data
rfe.fit(x_train, y_train)

# Get the selected features
selected_features = [column for column in x_train.columns if column  in x_train.columns[rfe.support_]]

```

```python
x3_train=x_train[selected_features]
x3_test=x_test[selected_features]
y3_train=y_train
y3_test=y_test
```

```python
# Anova Test
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(score_func=f_classif, k=15)  # Select top k features
selector.fit_transform(x_train, y_train)
selected_feature_indices = selector.get_support(indices=True)
selected_feature_indices 
x4_train=x_train.iloc[:,selected_feature_indices]
x4_test=x_test.iloc[:,selected_feature_indices]
y4_test=y_test
y4_train=y_train
```
